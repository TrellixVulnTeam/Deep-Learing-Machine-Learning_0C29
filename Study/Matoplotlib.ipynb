## Motaplotlib
### 시각화를 도와주는 라이브러리

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# sin 함수

# 데이터 준비
x = np.arange(0, 6, 0.1)    # 0에서 6까지 0.1 간격으로 증가
y = np.sin(x)

# 그래프 그리기
plt.plot(x, y)
plt.tick_params(axis='x', colors='white')    # x, y축 색상 변경
plt.tick_params(axis='y', colors='white')
plt.show()
```

```python
#cos 함수

# 데이터 준비
x = np.arange(0, 6, 0.1)    # 0에서 6까지 0.1 간격으로 증가
y1 = np.sin(x)
y2 = np.cos(x)

# 그래프 그리기
plt.plot(x, y1, label="sin")
plt.plot(x, y2, linestyle="--", label="cos")    # 선 스타일 점선
plt.tick_params(axis='x', colors='white')    # x, y축 색상 변경
plt.tick_params(axis='y', colors='white')
plt.xlabel("x")    # x축 이름
plt.ylabel("y")    # y축 이름
plt.title('sin & cos')    # 제목
plt.legend()
plt.show()
```

```python
# 이미지 표시하기

from matplotlib.image import imread

img = imread("img.jpg")    # 이미지 읽어오기

plt.imshow(img)
plt.tick_params(axis='x', colors='white')    # x, y축 색상 변경
plt.tick_params(axis='y', colors='white')
plt.show
```
